```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://vi.wikipedia.org/wiki/Danh_s%C3%A1ch_phim_%C4%91i%E1%BB%87n_%E1%BA%A3nh_c%C3%B3_doanh_thu_cao_nh%E1%BA%A5t_t%E1%BA%A1i_Vi%E1%BB%87t_Nam'
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/114.0 Safari/537.36"
}

website_url = requests.get(url, headers=headers).text
```

```python
soup = BeautifulSoup(website_url,'lxml')
#print(soup.prettify())
```

```python
soup
```

```python
s=soup.find('p')
print(s)
```

```python
s=soup.find_all('p')
for p in s:
    print(p)
```

```python
table = soup.find("table")
table 
```

```python
table = soup.find_all("table")[1]
```

```python
for film in table.find_all('tr'):
    print(film)
```

```python
table.find_all('tr')[1]
```

```python
elements = []
for film in table.find_all('tr'):
 # only find game names with links to their own pages
    try:
        name = film.find_all('td')[1].find('a').text
        link = 'https://vi.wikipedia.org' + film.find_all('td')[1].find('a').get('href')
    
    except: pass

    # find rest of cells
    else:
        cells = film.find_all('td') # mảng
        atts = []
        for i in range(len(cells)): # duyệt qua từng cột
            att = cells[i].find(text=True).strip()
            # "  abc  " -> "abc"
            atts.append(att)
        atts.insert(2,link)
        elements.append(atts)
     
     


print(elements)
```

```python
for x in elements:
    print(x) 
    print('----------------------------------------')
```

```python
for film in elements:
    print(film)
```

```python
features = ['Thứ Hạng','Phim','Link','Đạo diễn' ,'Doanh thu (tỷ VNĐ)', 'Diễn viên', 'Năm']
```

```python
import pandas as pd
df = pd.DataFrame(elements, columns=features)
df.head()
```

```python
df.set_index('Thứ Hạng', inplace=True)
```
