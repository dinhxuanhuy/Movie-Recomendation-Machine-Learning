```python
for idx,row in films.iterrows():
    url = row['Link']
    response = requests.get(url , headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    content = soup.find_all('div', class_='mw-heading mw-heading2')
    text = []
    for c in content :
        if c.text.startswith('Nội') or c.text.startswith('Plot'):
                for element in c.next_siblings:
                    if element.name and element.name.startswith('h') or element.name == 'div':
                        break
                    elif element.name == 'p':
                        text.append(element.text)
                break 
    films.at[idx, 'Content'] = ' '.join(text)
    print(f'Processed {idx+1}/{len(films)}: {row["Phim"]}')
```

```python
import re 
def clean_text(text):
    text = re.sub(r'\[\d+\]', '', text)  # Remove reference numbers like [1], [2], etc.
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces with a single space
    text = text.strip()  # Remove leading and trailing whitespace
    return text
```

```python
films['Content'] = films['Content'].apply(clean_text)
```
